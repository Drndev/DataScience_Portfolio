```python
# seperating the independant and dependant varible before test/train
X = data.drop(['PRICE'], axis = 1)
y = data['PRICE']
```

```python
# Further plotting of attributes which are more relevant to the Price of the house 
plt.figure(figsize=(20, 5))

features = ['LSTAT', 'TAX','RM', 'RAD']

target = data['PRICE']

for i, col in enumerate(features):
    plt.subplot(1, len(features) , i+1)
    x = data[col]
    y = target
    plt.scatter(x, y, marker='o')
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('PRICE')
```

```python
# Splitting to training and testing data

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.20, random_state = 66)
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
# Model prediction for the train dataset spilt
y_pred = lm.predict(X_train)
```

#### Evaluating the Model using R^2

The r-squared value is used so we can get a measure of the linear relationship between the  X and Y in the model. We look at the proportion of the variance in the dependent variable that will be predicted from the independent variable.

```python
print('R^2:',metrics.r2_score(y_train, y_pred))
```

```python
# Visualizing the differences between actual prices and predicted values
plt.scatter(y_train, y_pred)
plt.xlabel("Actual House Prices")
plt.ylabel("Predicted prices")
z = np.polyfit(y_train, y_pred, 1)
p = np.poly1d(z)
plt.plot(y_train,p(y_train),"r--")
plt.title("Linear Regressor Actual House Prices vs Predicted House Prices")
plt.show()
```

```python
# Model Evaluation

y_test_pred = lm.predict(X_test)

acc_linreg = metrics.r2_score(y_test, y_test_pred)
print('R^2:', acc_linreg)


```

#### + Working out the accuracy in a percentage (same as R^2 value onbtained above)

```python

y_test_pred = lm.predict(X_test)

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
accuracy = regressor.score(X_test,y_test)
print(accuracy*100,'%')
        
```

### Random Forest Regressor

```python

from sklearn.ensemble import RandomForestRegressor
reg = RandomForestRegressor()
reg.fit(X_train, y_train)
```

```python
#Evlauting RF 
y_pred = reg.predict(X_train)
print('R^2:',metrics.r2_score(y_train, y_pred))

```

```python
# Visualizing the differences between actual prices and predicted values
plt.scatter(y_train, y_pred)
plt.xlabel("Prices")
plt.ylabel("Predicted prices")
plt.title("Random Forest Actual House Prices vs Predicted House Prices")
z = np.polyfit(y_train, y_pred, 1)
p = np.poly1d(z)
plt.plot(y_train,p(y_train),"r--")
plt.show()
```

```python
# Predicting Test data with the model
y_test_pred = reg.predict(X_test)

# Model Evaluation

acc_rf = metrics.r2_score(y_test, y_test_pred)
print('R^2:', acc_rf)
```

```python
# Creating scaled set to be used in model to improve our results
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

### SVR

```python
# Import SVM Regressor
from sklearn import svm

# Create a SVM Regressor
reg = svm.SVR()
```

```python
# Train the model using the training sets 
reg.fit(X_train, y_train)
```

```python
# Model prediction on train data
y_pred = reg.predict(X_train)
# Model Evaluation
print('R^2:',metrics.r2_score(y_train, y_pred))
```

```python
# Visualizing the differences between actual prices and predicted values
plt.scatter(y_train, y_pred)
plt.xlabel("Prices")
plt.ylabel("Predicted prices")
plt.title("Suppourt Vector Regressor Actual House Prices vs Predicted House Prices")
z = np.polyfit(y_train, y_pred, 1)
p = np.poly1d(z)
plt.plot(y_train,p(y_train),"r--")
plt.show()
```

```python
# Predicting Test data with the model
y_test_pred = reg.predict(X_test)
# Model Evaluation
acc_svm = metrics.r2_score(y_test, y_test_pred)
print('R^2:', acc_svm)
```
