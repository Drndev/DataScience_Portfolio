```python
## PCA standarsing the data so that it can be correctly read for PCA 

scaled_data = preprocessing.scale(df.T)

pca=PCA()
pca.fit(scaled_data)
pca_data = pca.transform(scaled_data)
```

```python
per_var = np.round(pca.explained_variance_ratio_* 100, decimals=1)
labels = ['PC' + str(x) for x in range(1, len(per_var)+1)]
 
plt.bar(x=range(1,len(per_var)+1), height=per_var, tick_label=labels)
plt.ylabel('Percentage of Explained Variance')
plt.xlabel('Principal Component')
plt.title('Screen Plot')
plt.show()
```

```python
#the following code makes a plot  for PC1 and PC2 looking at all 31 features to see what features 
#are most imporant in each dimension 

pca_df = pd.DataFrame(pca_data, index=[list(df.columns[0:31])], columns=labels)
 
plt.scatter(pca_df.PC1, pca_df.PC2)
plt.title('My PCA Graph')
plt.xlabel('PC1 - {0}%'.format(per_var[0]))
plt.ylabel('PC2 - {0}%'.format(per_var[1]))
 
for sample in pca_df.index:
    plt.annotate(sample, (pca_df.PC1.loc[sample], pca_df.PC2.loc[sample]))
 
plt.show()
```

From the plot of the 2 dimesional PC1 and PC2 we can see almost 100% of the variance can be explained in the first dimension 
which can tell us that any varinace with the data can be explained by a single dimension

This will help with feature selection on the next step

# Gaussian Niave Bayes Classifier 

```python

from sklearn.metrics import confusion_matrix
gnb = GaussianNB()
gnb_scores = cross_val_score(gnb, X, y, cv=3, scoring='accuracy')
print(gnb_scores.mean())
```

# KNN

```python
# Decide what k should be for KNN
knn = KNeighborsClassifier()

k_range = list(range(1, 30))
leaf_size = list(range(1,30))
weight_options = ['uniform', 'distance']
algorithm = ['auto', 'ball_tree', 'kd_tree', 'brute']
param_grid = {'n_neighbors': k_range, 'leaf_size': leaf_size, 'weights': weight_options, 'algorithm': algorithm}
```

```python
rand_knn = RandomizedSearchCV(knn, param_grid, cv=3, scoring="accuracy", n_iter=100, random_state=45)
rand_knn.fit(X,y)
```

```python
print(rand_knn.best_score_)
```

# SVM Classifier 

```python
sv_clf = SVC(random_state=42)

param_grid = [
              {'C': [1, 10, 100, 1000], 
               'kernel': ['linear']
              },
              {'C': [1, 10, 100, 1000], 
               'gamma': [0.001, 0.0001], 
               'kernel': ['rbf']
              },
]
```

```python
grid_sv = GridSearchCV(sv_clf, param_grid, cv=3, scoring="accuracy")
grid_sv.fit(X,y)
```

```python
print(grid_sv.best_score_)
```

# Random Forest 

```python
from sklearn.ensemble import RandomForestClassifier
from scipy.stats import randint as sp_randint

rf_clf = RandomForestClassifier(random_state=42)

param_grid = {"max_depth": [3, None],
 
    "max_features":  sp_randint(1, 8),}
```

```python
rand_rf = RandomizedSearchCV(rf_clf, param_distributions=param_grid, n_iter=100, random_state=21, cv=3)
rand_rf.fit(X,y)
```

```python
print(rand_rf.best_score_)
```
